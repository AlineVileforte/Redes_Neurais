```python
from transformers import pipeline


sentiment_analyzer = pipeline("sentiment-analysis")


reviews = [
    "This movie was absolutely fantastic! I loved every minute of it.",
    "It was a complete waste of time. The plot was boring and the acting was terrible.",
    "Decent film, but nothing really stood out.",
    "I've never seen such a masterpiece. Truly groundbreaking!",
    "So disappointed. I expected more from this director."
]

print("--- Análise de Sentimento das Reviews ---")
for i, review in enumerate(reviews):
    result = sentiment_analyzer(review)
    label = result[0]['label']
    score = result[0]['score']
    print(f"\nReview {i+1}: '{review}'")
    print(f"Sentimento: {label} (Score: {score:.4f})")

results_batch = sentiment_analyzer(reviews)
print("\n--- Resultados em Lote ---")
for i, result in enumerate(results_batch):
    print(f"Review {i+1} -> Sentimento: {result['label']} (Score: {result['score']:.4f})")
```
